# 7203 data points for 10 epochs
Comparing the runs for RefinedModelArchitecture which considers the entire 7203 data points

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
MSE_uit = np.load('/media/disk1/prasad/codes/RefinedModelOutputs/CosmoUiT10/MSE.npy')
MSE_unet = np.load('/media/disk1/prasad/codes/RefinedModelOutputs/CosmoUNet10/MSE.npy')

R2_uit = np.load('/media/disk1/prasad/codes/RefinedModelOutputs/CosmoUiT10/R2.npy')
R2_unet = np.load('/media/disk1/prasad/codes/RefinedModelOutputs/CosmoUNet10/R2.npy')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5), facecolor='w', dpi=300)
plt.subplots_adjust(wspace=0.3)
axs[0].hist(MSE_uit, bins=100, color='steelblue', edgecolor='black')
axs[0].set_title("UiT", fontsize=14)
axs[0].set_xlabel("MSE", fontsize=12)
axs[0].set_ylabel("Frequency", fontsize=12)
axs[0].axvline(np.median(MSE_uit), color='orange', linestyle='--', linewidth=2, label=f'Median: {np.median(MSE_uit):.4f}')
axs[0].axvline(np.max(MSE_uit), color='red', linestyle='-', linewidth=2, label=f'Max: {np.max(MSE_uit):.4f}')
axs[0].legend(fontsize=10)

axs[1].hist(MSE_unet, bins=100, color='lightcoral', edgecolor='black')
axs[1].set_title("UNet", fontsize=14)
axs[1].set_xlabel("MSE", fontsize=12)
axs[1].set_ylabel("Frequency", fontsize=12)
axs[1].axvline(np.median(MSE_unet), color='orange', linestyle='--', linewidth=2, label=f'Median: {np.median(MSE_unet):.4f}')
axs[1].axvline(np.max(MSE_unet), color='red', linestyle='-', linewidth=2, label=f'Max: {np.max(MSE_unet):.4f}')
axs[1].legend(fontsize=10)

plt.suptitle("MSE Comparison", fontsize=16, fontweight='bold')
plt.show();
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5), facecolor='w', dpi=300)
plt.subplots_adjust(wspace=0.3)
axs[0].hist(R2_uit, bins=100, color='steelblue', edgecolor='black')
axs[0].set_title("UiT", fontsize=14)
axs[0].set_xlabel(r"$R^{2}$", fontsize=12)
axs[0].set_ylabel("Frequency", fontsize=12)
axs[0].axvline(np.median(R2_uit), color='orange', linestyle='--', linewidth=2, label=f'Median: {np.median(R2_uit):.4f}')
axs[0].axvline(np.max(R2_uit), color='red', linestyle='-', linewidth=2, label=f'Min: {np.min(R2_uit):.4f}')
axs[0].legend(fontsize=10)

axs[1].hist(R2_unet, bins=100, color='lightcoral', edgecolor='black')
axs[1].set_title("UNet", fontsize=14)
axs[1].set_xlabel(r"$R^{2}$", fontsize=12)
axs[1].set_ylabel("Frequency", fontsize=12)
axs[1].axvline(np.median(R2_unet), color='orange', linestyle='--', linewidth=2, label=f'Median: {np.median(R2_unet):.4f}')
axs[1].axvline(np.max(R2_unet), color='red', linestyle='-', linewidth=2, label=f'Min: {np.min(R2_unet):.4f}')
axs[1].legend(fontsize=10)

plt.suptitle(r"$R^{2}$ Comparison", fontsize=16, fontweight='bold')
plt.show();
```

### Possible Reason: Due to additional layer of transformer block it requires extra epochs for weights to converge.
